```python
import pygame
import datetime
import math

# инициализация Pygame
pygame.init()

# размеры экрана
screen_width = 500
screen_height = 500

# создание экрана
screen = pygame.display.set_mode((screen_width, screen_height))

# фоновый цвет
bg_color = (34, 37, 48)

# цвета для циферблата и стрелок
clock_color = (212, 212, 214)
hour_color = (252, 5, 58)
minute_color = (40, 40, 252)
second_color = (0, 255, 0)

# размеры и положения элементов циферблата
clock_radius = min(screen_width, screen_height) // 2 - 50
clock_center = (screen_width // 2, screen_height // 2)
hour_radius = clock_radius // 2
minute_radius = clock_radius * 3 // 4
second_radius = clock_radius - 10

# функция для отрисовки циферблата и цифр на нем
def draw_clock():
    # отрисовка круга циферблата
    pygame.draw.circle(screen, clock_color, clock_center, clock_radius, 5)

    # отрисовка цифр на циферблате
    for i in range(1, 13):
        angle = math.radians(i * 30 - 90)
        #Координаты каждой цифры вычисляем с использованием тригонометрических функций
        x = int(clock_center[0] + (clock_radius - 20) * math.cos(angle))
        y = int(clock_center[1] + (clock_radius - 20) * math.sin(angle))
        font = pygame.font.Font(None, 30)
        text = font.render(str(i), True, clock_color)
        text_rect = text.get_rect(center=(x, y))
        screen.blit(text, text_rect)

# функция для отрисовки стрелки
def draw_hand(color, angle, length):
    x = int(clock_center[0] + length * math.cos(angle))
    y = int(clock_center[1] + length * math.sin(angle))
    pygame.draw.line(screen, color, clock_center, (x, y), 6)
    
def draw_date():
    font = pygame.font.Font(None, 30)
    #Используем функцию strftime() для форматирования текущей даты.
    text = font.render(datetime.datetime.now().strftime('%d.%m.%Y'), True, clock_color)
    #вычисляет координаты для размещения текста в центре нижней части экрана
    text_rect = text.get_rect(center=(screen_width // 2, screen_height - 20))
    screen.blit(text, text_rect)

# главный цикл программы
while True:
    # обработка событий
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            quit()

    # очистка экрана
    screen.fill(bg_color)

    # отрисовка циферблата
    draw_clock()

    # получение текущего времени
    current_time = datetime.datetime.now().time()

    # вычисление угла для каждой стрелки
    hour_angle = math.radians((current_time.hour % 12) * 30 - 90)
    minute_angle = math.radians(current_time.minute * 6 - 90)
    second_angle = math.radians(current_time.second * 6 - 90)

    # отрисовка стрелок
    draw_hand(hour_color, hour_angle, hour_radius)
    draw_hand(minute_color, minute_angle, minute_radius)
    draw_hand(second_color, second_angle, second_radius)
    
    draw_date()

    # обновление экрана
    pygame.display.update()

    # задержка на 1 секунду
    pygame.time.delay(1000)
```

```python
import tkinter as tk
import time

def update_time():
    current_time = time.strftime('%H:%M:%S')
    clock_label.config(text=current_time)
    clock_label.after(1000, update_time)

root = tk.Tk()
root.title('Clock')

clock_label = tk.Label(root, font=('Arial', 200), bg='black', fg='white')
clock_label.pack(fill='both', expand=True)

update_time()

root.mainloop()
```
